<h2> Lets insert year_month column in 2nd position <h2>

<h2> Adding month column to the exisint DataFrame <h2>

```python
c2 = 'month'
v2 = df_new.invoice_date.dt.month
df_new.insert(loc = 3, column = c2, value = v2)
df_new.head()
```

<h2> We can get day of the week <h2>

<h2> In pandas, the day formate starts from 0 to 6
Apply +1 to make Monday = 1.....until Sunday = 7 <h2>

```python
c3 = 'day'
v3 = (df_new.invoice_date.dt.dayofweek)+1
df_new.insert(loc = 4, column = c3, value = v3)
df_new.head()
```
